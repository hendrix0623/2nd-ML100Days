```python
# Import 需要的套件
import os
import numpy as np
import pandas as pd
import statistics 
# 設定 data_path
dir_data = './documents/data/'
```

```python
f_app_train = os.path.join(dir_data, 'application_train.csv')
app_train = pd.read_csv(f_app_train)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

n = len(app_train['AMT_INCOME_TOTAL'])
avg = sum(app_train['AMT_INCOME_TOTAL'])/n

print('AMT_INCOME_TOTAL平均值:',avg)

sd =((sum((app_train['AMT_INCOME_TOTAL']-avg)**2)/n-1))**(1/2)

print('AMT_INCOME_TOTAL標準差:',sd)

print('AMT_INCOME_TOTAL(stdev):',statistics.stdev(app_train['AMT_INCOME_TOTAL']))

app_train = app_train.replace(np.nan,0) ##convert nan value to zero
app_train = app_train.groupby(['OCCUPATION_TYPE'])['AMT_INCOME_TOTAL'].mean()

app_train.plot.bar()
 
plt.show()
```
